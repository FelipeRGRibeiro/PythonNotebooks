```python
import seaborn as sn
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
def plot_cm(sklearn_cm, labels):
  array = sklearn_cm
  plt.figure(figsize = (8, 4.5))
  df_cm = pd.DataFrame(array, range(array.shape[0]), range(array.shape[1]))
  sn.set(font_scale = 1.4)
  p = sn.heatmap(df_cm, annot = True,
                 annot_kws = {"size": 14, "va":'bottom', "color":'k'},
                 fmt = 'd', linewidths = .5,
                 cmap = "Blues", cbar = False,
                 xticklabels = labels, yticklabels = labels)
  df_sum = df_cm.copy()
  df_sum = df_sum.sum(axis = 0)
  df_cm = df_cm.div(df_sum).to_numpy()
  p = sn.heatmap(df_cm, annot = True,
                 annot_kws = {"size": 14, "va":'top', "color":'k'},
                 fmt = '.2%', linewidths = .5,
                 cmap = "Blues", cbar = False,
                 xticklabels = labels, yticklabels = labels)
  p.set_yticklabels(p.get_yticklabels(), rotation = 30, fontsize = 14)
  p.set_xticklabels(p.get_xticklabels(), rotation = 30, fontsize = 14)
  plt.ylabel("Categoria\nverdadeira")
  plt.xlabel(f"Categoria Predita")  
  plt.show()
```

## Exemplo

```python
cm = np.array([[ 25,  29,  10,  40], [ 43,  93,  18, 140], [  0,  35,  14,  62], [ 62,  13,  30,  44]])
labels = ['Banana', 'Laranja', 'Maçã', 'Uva']
plot_cm(cm, labels)
```
