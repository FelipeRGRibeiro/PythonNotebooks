```python
import cv2
from PIL import Image
from urllib import request
from io import BytesIO
from google.colab.patches import cv2_imshow
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.preprocessing import image
```

##Definição do Data Augmentation

https://www.tensorflow.org/api_docs/python/tf/keras/preprocessing/image/ImageDataGenerator

```python
datagen = ImageDataGenerator(horizontal_flip = True,
                             vertical_flip = True,
                             rotation_range = 30,
                             zoom_range = 0.2,
                             width_shift_range = 0.1,
                             height_shift_range = 0.2,
                             shear_range = 0.2,
                             fill_mode='constant')
```

##Carregando imagem

```python
url = "https://upload.wikimedia.org/wikipedia/commons/thumb/d/dc/Buzz_Aldrin.jpg/479px-Buzz_Aldrin.jpg?20190717231121"
res = request.urlopen(url).read()
img = Image.open(BytesIO(res)).resize((224,224))
img_array = image.img_to_array(img)
cv2_imshow(cv2.cvtColor(img_array, cv2.COLOR_BGR2RGB))
```

##Imagem resultado do aumento de dados

```python
batches = datagen.flow(img_array.reshape(1,224,224,3))
cv2_imshow(cv2.cvtColor(next(batches)[0], cv2.COLOR_BGR2RGB))
```
